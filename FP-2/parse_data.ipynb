# parse_data.ipynb

This notebook parses the data files used for the FP-2 assignment. 

<br>
<br>

First let's read the attached data file:

```python
import pandas as pd

df0 = pd.read_csv('Real estate valuation data set.csv')

df0.describe()
```

<br>
<br>

The dependent and independent variables variables (DVs and IVs) that we are interested in are:

**DVs**:
- Price  (the house price of unit area)

**IVs**:
- Age ("X2 house age" column in the CSV file)
- From station ("X3 distance to the nearest MRT station" column in the CSV file)
- Number of convenience stores  ("X4 number of convenience stores" column in the CSV file)


<br>
<br>

Let's extract the relevant columns:

```python

df = df0[  ['Y house price of unit area', 'X2 house age', 'X3 distance to the nearest MRT station', 'X4 number of convenience stores']  ]

df.describe()
```

<br>
<br>

Next let's use the `rename` function to give the columns simpler variable names:

```python
df = df.rename( columns={'Y house price of unit area':'price','X2 house age':'age', 'X3 distance to the nearest MRT station':'from station', 'X4 number of convenience stores':'number of convenience stores'} )

df.describe()
```
