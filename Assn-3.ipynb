### Figure 1

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a = np.load( 'data-fig1.npy' )

plt.plot(a, 'o-', c='k')
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()
```

### Figure 2

```python
a = np.load( 'data-fig2.npy' )

data_red = a[:, 0]
data_green = a[:, 1]
data_blue = a[:, 2]
plt.plot(data_red, c='r', label='Red', linestyle='-')
plt.plot(data_green, c='g', label='Green', linestyle='-')
plt.plot(data_blue, c='b', label='Blue', linestyle='-')
plt.legend()
plt.xlabel("X value")
plt.ylabel("Y value")
plt.show()
```

### Figure 3

```python
a = np.load( 'data-fig3.npy' )

x_data = a[:, 0]
y_data = a[:, 1]
plt.scatter(x_data, y_data, s=200, c='b', alpha=0.3)
plt.xlabel('X value')
plt.ylabel('Y value')
plt.show()
```

### Figure 4

```python
a = np.load( 'data-fig4.npy' )

a_data = a[0]
b_data = a[1]
c_data = a[2]
d_data = a[3]
e_data = a[4]
heights=(a[0],a[1],a[2],a[3],a[4])
species=('A','B','C','D','E')
plt.bar(species, heights, color='darkgray')
plt.ylabel('Value')
plt.xlabel('Category')
plt.show()
```

### Figure 5

```python
a = np.load( 'data-fig5.npy' )

fig, ax = plt.subplots()
im = ax.imshow(a, cmap='RdBu', vmin=-1.0, vmax=1.0, interpolation='none', origin='lower', aspect='equal')
ax.set_ylim(0, 120)
cbar = fig.colorbar(im, extend='both',extendrect=True)
plt.show()
```
